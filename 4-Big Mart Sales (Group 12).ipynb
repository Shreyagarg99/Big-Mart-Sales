```python
# Checking different varieties of item fat content 

df['Item_Fat_Content'].value_counts(normalize = True)
df['Item_Fat_Content'].value_counts().plot.bar()
plt.title('Different varieties of fats in item in the store')
plt.xlabel('Fat')
plt.ylabel('Number of Items')
plt.show()
# Number of Low fat items present in the store are more as compared to items of other fat contents
```

```python
sns.pairplot(df)
```

```python
plt.hist(df['Item_MRP'], bins = 15, color = 'pink') 
plt.title('Target Variable')
plt.xlabel('Item MRP')
plt.ylabel('count')
plt.show()
```

```python
# Which year has maximum numer of outlets established?
plt.figure(figsize=(10,10))
sns.countplot(x="Outlet_Establishment_Year", data=df)
```

```python
# Checking fat content in different item types 
Item_Type = pd.crosstab(df['Item_Type'], df['Item_Fat_Content'])
Item_Type.div(Item_Type.sum(1).astype(float), axis=0).plot(kind="bar", stacked=True, figsize=(13, 13))

# Breakfast has the highest regular fat content and hard drinks, households, health and hygience have the highest low fat content
```

```python
# Barplot to determine the most expensive item
plt.figure(figsize=(20,6))
sns.barplot(x="Item_Type",y="Item_MRP",data=df)
plt.title=("Most Expensive Item")
plt.xlabel=("Type of Items")
plt.ylabel=("Price of the Item")
plt.show()
```

```python
# scatter plot to show the relation between Item visibility and Outlet sales
plt.figure(figsize=(12,8))
sns.scatterplot(x='Item_Visibility',y='Item_Outlet_Sales',hue='Item_Type',size='Item_Weight',data=df)
```

```python
# Barplot to determine if sales is affected by the size of /outlet
plt.figure(figsize =(18,6))
sns.boxplot("Item_Outlet_Sales", "Outlet_Size", data=df)
plt.show()
# It is not affected
```

```python
# Heat Map to show relation between Item type and Outlet Size with respect to item price
df=df.pivot_table(index='Item_Type',columns='Outlet_Size',values='Item_MRP')
sns.heatmap(df)
```

```python
# Violin plot to show the relationship between Item Fat content and Item weight
plt.figure(figsize=(12,6))
df=sns.violinplot(x="Item_Fat_Content", y="Item_Weight", data=df)
```

```python
# Conclusion
# Most of items are pricesd between 110-130
# Positive correlation between Outlet Sales and Item MRP
# Most expensive item is Dairy and household followed by Starchy Food
# Outlet visibility has no impact on outlet sales. Consumers tend to make the choice for a product
# Sales of Outlet size medium are the highest
# Maximum sales are in the year 1985 and ever since the sales have been constant
```
